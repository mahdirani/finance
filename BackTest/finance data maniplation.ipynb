```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("finance/gold-data/[2010-2020]XAUUSD-M5-NoSession.csv")
```

```python
df.head()
```

```python
# Converting Date format
df["Date"] = pd.to_datetime(df["Date"], format='%Y%m%d')
df
```

```python
# Combining Date and Time columns and set it as index
datetime = pd.to_datetime(df["Date"].astype(str) + " " + df["Time"])
df.insert(0, "DateTime", datetime)
df.pop("Date")
df.pop("Time")
df = df.set_index(['Datetime'])
df
```

```python
def Ichimoku(df):

  # Tenkan
  hi_in_9 = df["high"].rolling(window=9, min_periods=0).max()
  low_in_9 = df["low"].rolling(window=9, min_periods=0).min()
  df["Tenkan"] = (hi_in_9 + low_in_9) / 2

  # Kijun
  hi_in_26 = df["high"].rolling(window=26, min_periods=0).max()
  low_in_26 = df["low"].rolling(window=26, min_periods=0).min()
  df["kijuen"] = (hi_in_26 + low_in_26) / 2

  # Senkou Span A
  df["Senkou Span A"] = ((df["tenkan"]+df["kijuen"]) / 2)

  # Senkou Span B
  hi_in_52 = df["high"].rolling(window=52, min_periods=0).max()
  low_in_52 = df["low"].rolling(window=52, min_periods=0).min()
  df["senkou B"] = ((low_in_52 + hi_in_52) / 2)
```
